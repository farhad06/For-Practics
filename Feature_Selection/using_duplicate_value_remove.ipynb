### For only Numerical DataFrame

```python
import pandas as pd
```

```python
numerical_df=pd.read_csv("https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Santander_CS/train.csv")
```

```python
numerical_df.head()
```

```python
numerical_df.shape
```

```python
numerical_df=numerical_df.iloc[:,:-1]
```

```python
numerical_df.shape
```

```python
numerical_df_transpose=numerical_df.T
```

```python
numerical_df_transpose.shape
```

```python
numerical_df_transpose.head()
```

```python
#remove duplicate feature
duplicate_col=numerical_df_transpose[numerical_df_transpose.duplicated()].index.values
print(len(duplicate_col))
numerical_new_df=numerical_df.drop(columns=duplicate_col,axis=1)
```

```python
numerical_new_df.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('general_data.csv')
df.groupby(['Department'])['EmployeeCount'].count()
x=['Human Resources','Research & Development','Sales']
y=[189,2883,1338]
fig=plt.subplots(figsize=(10,4))
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.bar(x,y,width=0.4,color='g',label='Number of Employees',edgecolor='black')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('general_data.csv')
df.groupby(['Department'])['DistanceFromHome'].mean()
x=['Human Resources','Research & Development','Sales']
y=[8.253968,9.236212,9.230942]
fig=plt.subplots(figsize=(8,2))
plt.xlabel('Department')
plt.ylabel('Average Distance From Home')
plt.bar(x,y,width=0.4,color='b',label='Distance from Home',edgecolor='black')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('general_data.csv')
df.groupby('Department').EducationField.count()
rd=df[df.Department=='Research & Development']
rd.groupby('EducationField').EmployeeCount.count()
x=['Life Sciences','Medical','Other','Technical Degree']
y=[1320,1089,192,282]
fig=plt.subplots(figsize=(10,4))
plt.xlabel('Education Field')
plt.ylabel('Number of Employees')
plt.bar(x,y,width=0.2,color='r',label='Number of Employees',edgecolor='black')
plt.grid()
plt.legend()
plt.show()
```

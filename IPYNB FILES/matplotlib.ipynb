```python
import matplotlib.pyplot as plt
import numpy as np
```

## plot with line 

```python
x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show()
```

### plot without line

```python
plt.plot(x,y,"o")
plt.show()
```

 ## plot multiple point

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## plot default x

```python
# xpoints = np.array([1, 2, 6, 8])
# xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## plot using marker

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints,marker="o")
plt.show()
```

## plot using different color and different lines

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints,"o:r")
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints,"o--m")
plt.show()
```

## plot using marker size

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints,"o-.c",ms=20)
plt.show()
```

## plot using labels

```python
x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.xlabel("Year")
plt.ylabel("Age")
plt.show()
```

## add title

```python
x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.title("This is a simple Graph")
plt.xlabel("Year")
plt.ylabel("Age")
plt.show()
```

## use grid

```python
ypoints = np.array([3, 8, 1, 10])
plt.grid()
plt.plot(xpoints, ypoints,"o:r")
plt.show()
```

## Bar chart

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2.5, 15])

plt.bar(x,y)
plt.show()
```

## Horinzontal Bar Graph

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2.5, 15])

plt.barh(x,y)
plt.show()
```

## Bar chart using color

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2.5, 15])

plt.bar(x,y,color="red")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2.5, 15])

plt.barh(x,y,color="hotpink")
plt.show()
```

## bar chart using fixed width

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2.5, 15])

plt.bar(x,y,color="red",width=0.1)
plt.show()
```

## Pie chart

```python
y = np.array([30, 23, 25, 10])

plt.pie(y)
plt.show()
```

## adding labels

```python
y = np.array([30, 23, 25, 10])
label=["Farhad","Manoj","Imran","Azad"]

plt.pie(y,labels=label,startangle=90)
plt.show()
```

```python
y = np.array([30, 23, 25, 10])
label=["Farhad","Manoj","Imran","Azad"]
explode=[0.2,0.1,0.2,0.3]

plt.pie(y,labels=label,startangle=270,explode=explode)
plt.show()
```

```python
y = np.array([30, 23, 25, 10])
label=["Farhad","Manoj","Imran","Azad"]
explode=[0.2,0.1,0.2,0.3]

plt.pie(y,labels=label,startangle=90,explode=explode,shadow=True)
plt.show()
```

```python
y = np.array([30, 23, 25, 10])
label=["Farhad","Manoj","Imran","Azad"]
# explode=[0.2,0.1,0.2,0.3]
colors=["red","g","m","c"]

plt.pie(y,labels=label,startangle=90,colors=colors)
plt.show()
```

```python
y = np.array([30, 23, 25, 10])
label=["Farhad","Manoj","Imran","Azad"]
# explode=[0.2,0.1,0.2,0.3]
colors=["red","g","m","c"]


plt.pie(y,labels=label,startangle=90,colors=colors)
plt.legend()
plt.show()
```

```python
y = np.array([30, 23, 25, 10])
label=["Farhad","Manoj","Imran","Azad"]
# explode=[0.2,0.1,0.2,0.3]
colors=["red","g","m","c"]
plt.pie(y,labels=label,startangle=90,colors=colors)
plt.legend(title="Studends:")
plt.show()
```

## Sub plot

```python
x = np.array([5,8,10,12])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x,y)
x = np.array([8,2,5,6])
y = np.array([10,20,25,36])
plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([5,8,10,12])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 1, 1)
plt.plot(x,y)
x = np.array([8,2,5,6])
y = np.array([10,20,25,36])
plt.subplot(2,1, 2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([5,8,10,12])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.title("Graph-1")
plt.plot(x,y)
x = np.array([8,2,5,6])
y = np.array([10,20,25,36])
plt.subplot(1, 2, 2)
plt.title("Graph-2")
plt.plot(x,y)
plt.suptitle("This is an examle of Sub plot graph")
plt.show()
```
